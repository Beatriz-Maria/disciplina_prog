5. Use a função **calc_similarity** para calcular a similaridade entre as sequências (original e randomizadas).

```python
# Função que calcula a similaridade entre duas sequências
# Se o base/aminoácido for igual na posição soma 1 senão subtrai 1
def calc_similarity(seq1, seq2):
  similarity = 0
  for i in range(len(seq1)):
    if seq1[i] == seq2[i]:
      similarity += 1
    else:
      similarity -= 1
  return similarity
```
